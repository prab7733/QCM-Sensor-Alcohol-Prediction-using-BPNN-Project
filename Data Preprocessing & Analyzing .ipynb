# Rescaling Data

```python
import pandas, scipy, numpy
from sklearn.preprocessing import MinMaxScaler, StandardScaler, Normalizer, Binarizer, scale, OneHotEncoder, LabelEncoder
df=pandas.read_csv('QCM Sensor Alcohol Dataset\QCM10.csv',sep=';')
array=df.values

#Separating data into input and output components
x=array[:, 0:14]
y=array[:,0:14]
scaler=MinMaxScaler(feature_range=(0,1))
rescaledX=scaler.fit_transform(x)
numpy.set_printoptions(precision=3) #setting precision for the output
rescaledX[0:26,:]
```

# Standardizing Data

```python
scaler=StandardScaler().fit(x)
rescaledX=scaler.transform(x)
rescaledX[0:26,:]
```

# Normalizing Data

```python
scaler=Normalizer().fit(x)
normalizedX=scaler.transform(x)
normalizedX[0:25,:]
```

# Binarizing Data

```python
binarizer=Binarizer(threshold=0.1).fit(x)
binaryX=binarizer.transform(x)
binaryX[0:26,:]
```

# Mean Removal

```python
data_standardized=scale(df)
data_standardized.mean(axis=0)
```

# One Hot Encoding

```python
encoder=OneHotEncoder()
encoder.fit([[0,1,6,2],[1,5,3,5],[2,4,2,7],[1,0,4,2]])
```

```python
encoder.transform([[2,4,3,4]]).toarray()
```

# Label Encoding

```python
label_encoder=LabelEncoder()
input_classes=['1-Octanol','1-propanol','2-butanol','2-propanol','1-isobutanol']
label_encoder.fit(input_classes)
```

```python
for i, item in enumerate(label_encoder.classes_):
    print(item,'-->',i)
```

```python
labels=['1-propanol','2-propanol','1-Octanol']
label_encoder.transform(labels)
```

```python
label_encoder.inverse_transform(label_encoder.transform(labels))
```
